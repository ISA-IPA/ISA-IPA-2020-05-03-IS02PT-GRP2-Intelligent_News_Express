```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import io
```

```python
text = ""
with io.open('msft.txt', 'r', encoding='utf-8') as f:
    text = ''.join(f.readlines())
```

```python
wc = WordCloud(background_color="white")
```

```python
wc.generate(text)
```

```python
wc.to_file('cloud_word_output_1.png')
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
custom_mask = np.array(Image.open("Picture 1.png"))
wc = WordCloud(background_color="white", mask=custom_mask, contour_width=3, contour_color='white')

wc.generate(text)
wc.to_file('cloud_word_output_2.png')
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
